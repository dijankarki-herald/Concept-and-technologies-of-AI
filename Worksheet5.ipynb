```python
import pandas as pd
import numpy as np

# Redefine the split function with corrections
def split(X, Y, test_split=0.2, seed=42):
  np.random.seed(seed)
  indices = np.arange(X.shape[0])
  np.random.shuffle(indices)

  split_sizes = int(len(X) * test_split)
  train_indices = indices[split_sizes:]
  test_indices = indices[:split_sizes]

  X_train, X_test = X[train_indices], X[test_indices]
  Y_train, Y_test = Y[train_indices], Y[test_indices]

  return X_train, X_test, Y_train, Y_test

# Redefine cost_function (as it was previously defined in another cell)
def cost_function(X, Y, W):
  n = len(Y)
  y_pred = np.dot(X, W)
  error = Y - y_pred
  squares = np.square(error)
  total = np.sum(squares)
  result = total / n
  return result

# Redefine rmse (as it was previously defined in another cell)
def rmse(Y, Y_pred):
  errors = np.square(Y - Y_pred)
  averages = np.mean(errors)
  RMSE = np.sqrt(averages)
  return RMSE

# Redefine r2 (as it was previously defined in another cell)
def r2(Y, Y_pred):
  y_mean = np.mean(Y)
  ssr = np.square(Y - Y_pred)
  ssr_total = np.sum(ssr)
  sst = np.square(Y - y_mean)
  sst_total = np.sum(sst)
  rsquare = 1 - (ssr_total / sst_total)
  return rsquare

# Redefine gradient_descent with corrections
def gradient_descent(X, Y, W, alpha, iterations):
  m = len(Y)
  cost = []
  w_update = W.copy()

  # Ensure Y is 1D for consistency with other functions like rmse, r2, and the original test cases
  if Y.ndim > 1 and Y.shape[1] == 1:
      Y = Y.flatten()

  for i in range(iterations):
    y_prediction = np.dot(X, w_update)
    losses = y_prediction - Y # Now Y and y_prediction should have compatible shapes

    # Correct dw calculation: remove np.sum which was incorrectly aggregating gradients
    dw = (2/m) * np.dot(X.T, losses)

    w_update = w_update - alpha * dw
    costs = cost_function(X, Y, w_update)
    cost.append(costs)
    rsquare = r2(Y, y_prediction)
    rmse_values = rmse(Y, y_prediction)
    # print(f"for iteration {i} the cost function is {costs} r2= {rsquare} rmse= {rmse_values}")
  return w_update, cost


def main():
    # Step 1: Load dataset
    data = pd.read_csv("/content/drive/MyDrive/student.csv")

    # Step 2: Features and target
    X = data[["Math", "Reading"]].values
    Y = data["Writing"].values # Y will be a 1D array

    # Step 3: Train/test split
    # Now calls the corrected split function defined in this cell
    X_train, X_test, Y_train, Y_test = split(X, Y, test_split=0.2, seed=42)
    # Expected shapes: X_train (800, 2), X_test (200, 2), Y_train (800,), Y_test (200,)

    # Step 4: Initialize weights
    W = np.zeros(X_train.shape[1])
    alpha = 0.00001
    iterations = 1000

    # Step 5: Gradient Descent
    W_optimal, cost_history = gradient_descent(X_train, Y_train, W, alpha, iterations)

    # Step 6: Predictions on test set
    Y_pred = np.dot(X_test, W_optimal)

    # Step 7: Evaluation
    model_rmse = rmse(Y_test, Y_pred)
    model_r2 = r2(Y_test, Y_pred)

    # Step 8: Results
    print("\nFinal Weights:", W_optimal)
    print("Cost History (First 10):", cost_history[:10])
    print("RMSE on Test Set:", model_rmse)
    print("R-Squared on Test Set:", model_r2)

# Execute
if __name__ == "__main__":
    main()
```
